```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.age = le.fit_transform(df.age)
df.income = le.fit_transform(df.income)
df.student = le.fit_transform(df.student)
df.Credit_rating = le.fit_transform(df.Credit_rating)
df.Buys_computer = le.fit_transform(df.Buys_computer)
```

```python
df
```

```python
df = df.drop('Unnamed: 0',axis =1)
df
```

```python
x = df.drop('Buys_computer',axis=1).values
y = df.Buys_computer.values
print(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#dtc = DecisionTreeClassifier(criterion='entropy')
dtc = DecisionTreeClassifier(criterion='gini')
```

```python
dtc.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,dtc.predict(x_test))
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
plot_tree(dtc,feature_names=['age','income','student','Credit_rating'],class_names=['yes','no'],filled=True)
```
