```python
from tensorflow.keras.preprocessing import image
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import os
```

```python
dir = 'Augmented Image'
os.mkdir(dir)
```

```python
data_config = ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.2,
    height_shift_range=0.2,
    shear_range = 0.2,
    zoom_range =0.2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)
```

```python
# Reading image
img_path = '/content/images.jpeg'
img = image.load_img(img_path,target_size=(224,224))
```

```python
# Reshaping image
x = image.img_to_array(img)
x = x.reshape((1,)+x.shape)
```

```python
for batch,batch_id in zip(data_config.flow(x,batch_size=1),range(10)):
  aug_img = image.array_to_img(batch[0])
  aug_img.save(os.path.join(dir,f'aug_{batch_id}.jpg'))
```
