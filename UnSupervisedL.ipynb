```python
data.set_index('Country',inplace=True)
```

```python
data['Language']= data['Language'].map({"English":0,
                                        'French':1,
                                       'German':2})
```

```python
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.clustermap(data)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6).fit(data)
```

```python
# inertia_ == within the cluster interia value should be minimal amount.
# cluster value should be less than total amount of data
```

```python
# to find the wccs(within the cluster sum of squares)

model.inertia_
```

```python
model1 = KMeans(n_clusters = 5).fit(data)
```

```python
model1.inertia_
```

```python
wccs = []
for i in range(1,7):
  model = KMeans(n_clusters= i).fit(data)
  wccs.append(model.inertia_)
```

```python
wccs
```

```python
plt.plot(range(1,7),wccs)
plt.show()
# Elbow plotting
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
x = data.data
x
```

```python
kmeans = KMeans(n_clusters= 3).fit(x)
```

```python
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print(cluster_labels)
print(centroids)
```

```python
plt.scatter(x[:,0],x[:,1],c = cluster_labels,cmap= 'viridis')
plt.scatter(centroids[:,0],centroids[:,1],c = 'red',marker= x)
```
