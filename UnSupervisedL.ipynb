```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6).fit(data)
```

```python
# inertia_ == within the cluster interia value should be minimal amount.
# cluster value should be less than total amount of data
```

```python
# to find the wccs(within the cluster sum of squares)

model.inertia_
```

```python
model1 = KMeans(n_clusters = 5).fit(data)
```

```python
model1.inertia_
```

```python
wccs = []
for i in range(1,7):
  model = KMeans(n_clusters= i).fit(data)
  wccs.append(model.inertia_)
```

```python
wccs
```

```python
plt.plot(range(1,7),wccs)
plt.show()
# Elbow plotting
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
x = data.data
x
```

```python
kmeans = KMeans(n_clusters= 3).fit(x)
```

```python
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print(cluster_labels)
print(centroids)
```

```python
x[:,0],x[:,1]
```

```python
centroids[:,0],centroids[:,1]
```

```python
plt.scatter(x[:,0],x[:,1],c=cluster_labels,cmap='viridis')
plt.scatter(centroids[:,0],centroids[:,1],c='red',marker='x',s=100)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering')
plt.show()
```

```python
df=pd.DataFrame(x,columns=['Sepal Length','Sepal Width','Petal Length','Petal Width'])
df['k_target']=kmeans.labels_
df['target']=data.target
df
```

```python
from sklearn.metrics import silhouette_score
```

```python
ss=[]
for i in range(2,7):
  kmeans=KMeans(n_clusters=i).fit(x)
  ss.append(silhouette_score(x,kmeans.labels_))
```

```python
ss
```

```python
plt.plot(range(2,7),ss)
plt.show()
```
